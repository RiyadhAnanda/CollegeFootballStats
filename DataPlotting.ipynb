```python
# Imports
import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dataset Upload
dataset = pd.read_csv("CFB.csv")
# Select indexes 1-88 (remove empty rows) and set equal to dataset
dataset = dataset.iloc[1:88]
# Call .sample which will shuffle/randomize the rows/indexes, frac = 1 will return 100% of the dataframe
randomized = dataset.sample(frac = 1)
# Print first 5 indexes/rows to confirm 
#randomized.head()
```

```python
# Print info on data
# randomized.info()
```

```python
# Replace #DIV/0! (non-values) with 'NaN' or NULL values
randomized = randomized.replace('#DIV/0!', 'NaN')
# randomized.tail()
```

```python
#Print correlation of dataset
randomized.corr(numeric_only=True)
```

```python
# Relationship Between Conference and Total Touches/TD variables
sns.relplot(data=randomized,x='Year', y = 'RBR', hue='Binary All American')
plt.savefig('RelationalPlot.png')
```

```python
sns.pairplot (randomized, hue = 'Binary All American')
plt.savefig('AllStats.png')
```
